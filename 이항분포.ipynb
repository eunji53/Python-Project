````python
# 1Îã®Í≥Ñ: Ïù¥Ìï≠Î∂ÑÌè¨Ïùò ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
n = 10        # Ï¥ù ÏãúÎèÑ ÌöüÏàò
p = 0.5       # ÏÑ±Í≥µ ÌôïÎ•†
num_samples = 10000  # ÏÉùÏÑ±Ìï† ÎÇúÏàò Í∞úÏàò

# 2Îã®Í≥Ñ: Ïù¥Ìï≠Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî ÎÇúÏàò ÏÉùÏÑ±
samples = np.random.binomial(n, p, num_samples)
samples
````

````python
# 3Îã®Í≥Ñ: ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú Í≤∞Í≥º ÏãúÍ∞ÅÌôî
plt.hist(samples, bins = 'auto', density=True, 
         alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('ÏÑ±Í≥µ ÌöüÏàò k')
plt.ylabel('ÏÉÅÎåÄ ÎπàÎèÑ')
plt.title(f'Ïù¥Ìï≠Î∂ÑÌè¨ ÏãúÎÆ¨Î†àÏù¥ÏÖò (n={n}, p={p}, ÏÉòÌîå={num_samples}Í∞ú)')
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Ïù¥Ìï≠Î∂ÑÌè¨Ïùò ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
n = 10        # Ï¥ù ÏãúÎèÑ ÌöüÏàò (Ïòà: ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞ 10Ìöå)
p = 0.5       # Í∞Å ÏãúÎèÑÏóêÏÑú ÏÑ±Í≥µ ÌôïÎ•†

# k Í∞í (0Î∂ÄÌÑ∞ nÍπåÏßÄ)
k_values = np.arange(0, n+1)

# Í∞Å kÏóê ÎåÄÌïú ÎàÑÏ†ÅÌôïÎ•†(CDF) Í≥ÑÏÇ∞
cdf_values = binom.cdf(k_values, n, p)
cdf_values
````

````python
for i, j in zip(k_values, cdf_values):
    print(f"ÏïûÎ©¥Ïù¥ {i}Î≤à Ïù¥ÌïòÎ°ú ÎÇòÏò¨ ÌôïÎ•† : {j * 100}%")
````

````python
# ÎàÑÏ†ÅÌôïÎ•†ÏßàÎüâÌï®ÏàòÎ•º ÏãúÍ∞ÅÌôî (Í≥ÑÎã® Í∑∏ÎûòÌîÑ)
plt.step(k_values, cdf_values, label='ÎàÑÏ†Å ÌôïÎ•†')
plt.xlabel('ÏÑ±Í≥µ ÌöüÏàò k')
plt.ylabel('ÎàÑÏ†Å ÌôïÎ•† P(X ‚â§ k)')
plt.title(f'Ïù¥Ìï≠Î∂ÑÌè¨ ÎàÑÏ†Å ÌôïÎ•† ÏßàÎüâ Ìï®Ïàò (n={n}, p={p})')
plt.grid(True)
plt.legend()
plt.show()
````

2. Ìïú Í≥µÏû•ÏóêÏÑú ÏÉùÏÇ∞ÎêòÎäî Ï†úÌíàÏùò Í≤∞Ìï®Î•†Ïù¥ 0.02Ïùº Îïå, 100Í∞úÏùò Ï†úÌíàÏùÑ Í≤ÄÏÇ¨ÌïòÏó¨ Í≤∞Ìï®Ïù¥ 5Í∞ú Ïù¥ÏÉÅÏùº ÌôïÎ•†ÏùÑ Íµ¨ÌïòÏãúÏò§.

````python
n2 = 100        # Í≤ÄÏÇ¨Ìïú Ï†úÌíàÏùò Í∞úÏàò
p2 = 0.02       # Í≤∞Ìï®Î•†
# Í≤∞Ìï®Ïù¥ 5Í∞ú Ïù¥ÏÉÅÏùº ÌôïÎ•† = 1 - P(Í≤∞Ìï®Ïù¥ 4Í∞ú Ïù¥Ìïò)
prob2 = 1 - binom.cdf(4, n2, p2)
print("Î¨∏Ï†ú 2: 100Í∞ú Ï†úÌíà Ï§ë Í≤∞Ìï®Ïù¥ 5Í∞ú Ïù¥ÏÉÅÏùº ÌôïÎ•† =", prob2)
````
